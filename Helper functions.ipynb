### CASIA 1.0
Size: 200 MB

General info: All the images are divided into 3 folders: Au (Authentic image) and Modified tp and Original tp (Tampered images). Tampered images are further divided into copy-move forgery and image splicing.

Running the below cells will move them into 2 folders: au (Authentic) and tp (Tampered)

```python
dataset = 'CASIA 1.0 dataset'
images_path_au = base_path + '/' + dataset + '/CASIA 1.0 dataset' + '/Au' + '/Au'
images_path_tp1 = base_path + '/' + dataset + '/CASIA 1.0 dataset' + '/Modified Tp'
images_path_tp2 = base_path + '/' + dataset + '/CASIA 1.0 dataset' + '/Original Tp'

new_images_path_au = base_path + '/' + dataset + '/au'
new_images_path_tp = base_path + '/' + dataset + '/tp'
```

```python
def move_casia_v10():
    # moving authentic images
    if not os.path.exists(new_images_path_au):
        os.mkdir(new_images_path_au)
    
    for _, __, images_folder in os.walk(images_path_au):
        for image_name in images_folder:
            shutil.copy(images_path_au + '/' + image_name, new_images_path_au + '/' + image_name)
    
    # moving tampered images
    if not os.path.exists(new_images_path_tp):
        os.mkdir(new_images_path_tp)
    
    for _, __, images_folder in os.walk(images_path_tp1):
        for images_name in images_folder:
            path = _ + '/' + images_name
            
            shutil.copy(path, new_images_path_tp + '/' + images_name)
            
# move_casia_v10()
```

### MICC-F2000
Size: 2 GB

General info: Tampered images contain tamp in their name. All the images are present in a single folder

Running the below cells will move them into 2 folders: au (Authentic) and tp (Tampered)

```python
dataset = 'MICC-F2000'
images_path = base_path + '/' + dataset + '/MICC-F2000'

new_images_path_au = base_path + '/' + dataset + '/au'
new_images_path_tp = base_path + '/' + dataset + '/tp'
```

```python
def move_MICC_F2000():
    if not os.path.exists(new_images_path_au):
        os.mkdir(new_images_path_au)
        
    if not os.path.exists(new_images_path_tp):
        os.mkdir(new_images_path_tp)
    
    for _, __, images_folder in os.walk(images_path):
        for image_name in images_folder:
            path = images_path + '/' + image_name

            if 'tamp' in image_name:
                # moving tampered images
                shutil.copy(path, new_images_path_tp + '/' + image_name)
            else:
                # moving authentic images
                shutil.copy(path, new_images_path_au + '/' + image_name)
    
# move_MICC_F2000()
```

### Defacto - Copy Move
Size: 10 GB

```python
dataset = base_path + 'Defacto_CopyMove'
img_dir = dataset + '/copymove_img/img/'
mask_dir = dataset + '/copymove_annotations/probe_mask'

new_img_dir = dataset + '/defacto_bootstrap/img'
new_mask_dir = dataset + '/defacto_bootstrap/mask'

print(f'{img_dir=}, {mask_dir=}')
```

```python
def move_Defacto(limit=6000):
    if not os.path.exists(new_img_dir):
        os.makedirs(new_img_dir)
    
    if not os.path.exists(new_mask_dir):
        os.makedirs(new_mask_dir)
    
    for _, _, imgs in os.walk(img_dir):
        for img in tqdm(imgs[:limit]):
            img_path = img_dir + img
#             print(img_path)
            shutil.copy(img_path, new_img_dir + '/' + img)
    
    for _, _, imgs in os.walk(mask_dir):
        for img in tqdm(imgs[:limit]):
            img_path = mask_dir + '/' + img
#             print(img_path)
            shutil.copy(img_path, new_mask_dir + '/' + img)
    
    images = os.listdir(new_img_dir)
    masks = os.listdir(new_mask_dir)
    
    for image, mask in zip(images, masks):
        assert image.split('.')[0] == mask.split('.')[0]
        
        
move_Defacto()
```

### Data augmentation

```python
import tensorflow as tf
import numpy as np
import cv2
import random
from tqdm import tqdm
import matplotlib.pyplot as plt
from PIL import Image
import os

def augment(path):
    for name in tqdm(os.listdir(path)):
        try:
            image = cv2.imread(path + name)
            aug_images = tf.image.random_flip_left_right(image)
            aug_images = tf.image.random_flip_up_down(aug_images)
    #         aug_images = tf.image.random_brightness(aug_images, 0.2)
    #         aug_images = tf.image.random_crop(aug_images, size=(image.shape[0]-random.randint(5, 100), image.shape[1]-random.randint(5, 100), 3))
            aug_images = np.asarray(tf.image.rot90(aug_images))[:, :, ::-1]
            pil = Image.fromarray(aug_images, "RGB")
            pil.save(path + 'aug_' + name)
        except:
            print(name)
#         pil.show()
        

        
def remove_aug(path):
    for image in tqdm(os.listdir(path)):
        if 'aug' in image:
            os.remove(path + image)


# augment('../Data/CASIA2/tp/')
# augment('../Data/CASIA 1.0 dataset/au/')
remove_aug('../Data/CASIA2/tp/')
# remove_aug('../Data/CASIA 1.0 dataset/au/')
```
