```python
import os
from tqdm import tqdm
from PIL import Image, ImageChops, ImageEnhance
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def calculate_ela(image_path, save_path, quality=80):
    original_image = Image.open(image_path)
    original_image.save('temp.jpg', 'JPEG', quality=quality)

    # Open the re-saved image
    resaved_image = Image.open('temp.jpg')

    # Calculate the ELA
    ela_image = ImageChops.difference(original_image, resaved_image)
    extrema = ela_image.getextrema()
    max_diff = max([ex[1] for ex in extrema])
    scale = 255.0 / max_diff
    
    os.remove('temp.jpg')

    ela_image = ImageEnhance.Brightness(ela_image).enhance(scale)
    ela_image.save(save_path)
```

```python
# path = '../Data/Defacto_Splicing/defacto_bootstrap/img'
# ela_path = '../Data/Defacto_Splicing/defacto_bootstrap/ela/'
path = '../Data/CASIA2.0_revised_corrected/casia/Tp'
ela_path = '../Data/CASIA2.0_revised_corrected/casia/ela'

if not os.path.exists(ela_path):
        os.makedirs(ela_path)

# if not os.path.exists(ela_path + '/tp'):
#         os.makedirs(ela_path + '/tp')

print(1)
for _, __, ___ in os.walk(path):
    print(1)
    for files in tqdm(___):
        try:
            image_path = os.path.join(_, files)
            save_path = ela_path
            calculate_ela(image_path)
        except Exception as e:
            print(image_path, e)
#         break
```
