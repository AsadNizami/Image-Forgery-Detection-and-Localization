```python
import os
import pathlib
import random
import numpy as np
import matplotlib.pyplot as plt
 
import zipfile
import requests
import glob as glob
 
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Dropout, Flatten
from tensorflow.keras.utils import image_dataset_from_directory
 
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter)
from dataclasses import dataclass

from PIL import Image
```

### Sets random seeds for reproducibility

```python
SEED_VALUE = 41
dims = (224, 224)
 
random.seed(SEED_VALUE)
np.random.seed(SEED_VALUE)
tf.random.set_seed(SEED_VALUE)

tf.config.run_functions_eagerly(True)
```

### Loads train and validation image datasets

```python
path = '../../Dissertation/Data/MICC-F2000_ELA_80'
training_ds, val_ds = tf.keras.utils.image_dataset_from_directory(
    path,
    image_size=dims,
    batch_size=12,
    validation_split=0.25,
    subset='both', 
    seed=SEED_VALUE
)
```

```python
training_ds.class_names
```

```python
plt.figure(figsize=(18,10))
 
num_rows = 3
num_cols = 4
class_names = training_ds.class_names
for image_batch, labels_batch in training_ds.take(1):  
 
    for i in range(num_rows*num_cols):
        ax = plt.subplot(num_rows, num_cols, i + 1)
        plt.imshow(image_batch[i].numpy().astype("uint8"))
        truth_idx = labels_batch[i].numpy()
        plt.title(class_names[truth_idx])
        plt.axis("off")
```

```python
def evaluate(training_results, title):
#     print(training_results.history.items())
    
#     print("Params:", training_results.params)
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9, 4))
    ax1.plot(training_results.history['accuracy'], label='train accuracy')
    ax1.plot(training_results.history['val_accuracy'], label='val accuracy')
    ax1.legend(['Train', 'Validation'])
    ax1.title.set_text('Accuracy')
    ax1.set_xlabel('epoch')

    ax2.plot(training_results.history['loss'], label='train loss')
    ax2.plot(training_results.history['val_loss'], label='val loss')
    ax2.legend(['Train', 'Validation'])
    ax2.title.set_text('Loss')
    ax2.set_xlabel('epoch')
    plt.tight_layout()
    fig.suptitle(title)
```

### Loads EfficientNetV2B1 conv base with pretrained imagenet weights

```python
input_shape = (224, 224, 3)
effnet_v2_conv_base = tf.keras.applications.efficientnet_v2.EfficientNetV2B1(input_shape=input_shape,
                                                    include_top=False,
                                                    weights='imagenet',
                                                   )
```

### Loops through layers freezing all but last 5 conv layers

```python
effnet_v2_conv_base.trainable = True
LAYERS_FINE_TUNE = 5
num_layers_fine_tune = LAYERS_FINE_TUNE
num_layers = len(effnet_v2_conv_base.layers)
 
for model_layer in effnet_v2_conv_base.layers[:num_layers - num_layers_fine_tune]:
    print(f"FREEZING LAYER: {model_layer}")
    model_layer.trainable = False
 
print("\n")
print(f"Configured to fine tune the last", num_layers_fine_tune, "convolutional layers...")
print("\n")
     
print(effnet_v2_conv_base.summary())
```

### Model architecture

```python
inputs = tf.keras.Input(shape=input_shape)

x = tf.keras.applications.efficientnet_v2.preprocess_input(inputs)

x = effnet_v2_conv_base(x)

x = layers.GlobalAveragePooling2D()(x)
```

```python
outputs = layers.Dense(2, activation='softmax')(x)
model_finetune = keras.Model(inputs, outputs)
 
print(model_finetune.summary())
```

```python
reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.001)

model_finetune.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=2e-3),
                    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
                    metrics=['accuracy'],
                   )
```

```python
training_results = model_finetune.fit(training_ds,
                                      epochs=20,
                                      validation_data=val_ds,
                                      callbacks=[reduce_lr]
                                           )
```

### Result evaluation

```python
evaluate(training_results, 'MICC-F2000')
```
